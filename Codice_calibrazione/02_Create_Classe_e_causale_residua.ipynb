## Create the SAE ATECO normalized for the periemter

```python
df_perimeter = pd.read_excel(os.path.join(output_dir, "NETECH_tutti_scen_stato_v3.xlsx"))
df_perimeter.shape
```

```python
df_perimeter["ATECO_CODE"] = df_perimeter["D22 ATECO"].str.replace("\xa0", " ", regex=True).str[:3]
df_perimeter["ATECO_CODE"] = df_perimeter["ATECO_CODE"].fillna("999")
df_perimeter[['D22 ATECO','ATECO_CODE']].drop_duplicates()
```

```python
test = df_perimeter[df_perimeter['DESCRIZIONE'] == "Versamento giornaliero di importo elevato per la causale ed il soggetto"]
test[['D22 ATECO','ATECO_CODE']].drop_duplicates()
```

```python
df_perimeter["SAE_CODE"] = df_perimeter["D21 SAE"].str.replace("\xa0", " ", regex=True).str[:3]
df_perimeter[['D21 SAE','SAE_CODE']].drop_duplicates()
```

```python
df_perimeter["SAE_ATECO_CODE"] = (df_perimeter["SAE_CODE"].astype(str) + df_perimeter["ATECO_CODE"].astype(str))

lengths = df_perimeter["SAE_ATECO_CODE"].astype(str).str.len()
print(set(lengths))
df_perimeter[['SAE_CODE', 'ATECO_CODE', 'SAE_ATECO_CODE']].drop_duplicates()
```

Mapping the residual classes

```python
mask = df_joined["RISCHIO CONTANTE"].isna()
df_joined.loc[mask, "ATECO_CODE"] = "999"
df_joined.loc[mask, "RES_ATECO"] = '1'

print(df_joined[mask].shape)

df_joined = df_joined.rename(columns={"RISCHIO CONTANTE": 'Classe_old'})

df_joined.loc[mask, "SAE_ATECO_CODE"] = (df_joined.loc[mask, "SAE_CODE"].astype(str) + df_joined.loc[mask, "ATECO_CODE"].astype(str))

df_joined = df_joined.merge(
    df_classe[["SAE_ATECO_CODE", "RISCHIO CONTANTE"]].drop_duplicates(),
    how="left",
    on="SAE_ATECO_CODE")

df_joined = df_joined.rename(columns={"RISCHIO CONTANTE": 'Classe'})
print(df_joined.shape)
```

```python
mask_ = df_joined["RES_ATECO"] == '1'
df_joined.loc[~mask_,'Classe_old'].value_counts()
```

```python
mask_ = df_joined["RES_ATECO"] == '1'
df_joined.loc[~mask_,'Classe'].value_counts()
```

Replace causale sportello and VV with SPECIAL

```python
Special_Causale = ['451', '452', '14M', '56', 'ES', 'Y11', 'Y12', 'Y13', 'AFI']
df_joined["CAUSALE ANALITICA"] = df_joined["CAUSALE ANALITICA"].replace(Special_Causale, "SPECIAL")
df_joined.shape
```

```python
output_file = os.path.join(output_dir, "NETECH_CAMPIONE_CALIB.xlsx")
# Export DataFrame to Excel
df_joined.to_excel(output_file, index=False)
```
