```python
import json, re
from pathlib import Path

root = Path(r"C:\percorso\della\cartella")  # <-- cambi qui
pat = re.compile(r"""[^\s"'<>]+\.xlsx""", re.IGNORECASE)

hits = {}  # xlsx -> set(ipynb)

for nb in root.rglob("*.ipynb"):
    try:
        data = json.loads(nb.read_text(encoding="utf-8"))
    except Exception:
        continue
    for cell in data.get("cells", []):
        src = "".join(cell.get("source", []))
        for m in pat.findall(src):
            hits.setdefault(m, set()).add(str(nb))

for xlsx in sorted(hits):
    print(xlsx)
    for nb in sorted(hits[xlsx]):
        print("  -", nb)
```
